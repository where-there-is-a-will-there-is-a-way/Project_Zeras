```python
import numpy as np

t = np.arange(0, 2*np.pi, 0.01)
# numpy를 이용하여 값을 만듬!

plt.figure(figsize=(10,6))
# X값을 만들어줌!

plt.plot(t, np.sin(t))
# t의 값을 가지고 Y값을 만들어줌!

plt.grid()
plt.title('sine')
plt.xlabel('sec')
plt.ylabel('sin')
# title, xlabel, ylabel로 상단, 하단, 좌측 이름을 바꾸어줌!
plt.show()
```

```python
t = np.arange(0, 2*np.pi, 0.01)

plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), lw=3, label='sin')
plt.plot(t, np.cos(t), 'r', label='cos')
plt.grid()
plt.legend()
plt.title('sine')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```
